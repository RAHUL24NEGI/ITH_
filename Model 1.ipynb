```python
import os
import pandas as pd
import torch
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, Dataset
from torchvision import models
import torch.nn as nn
import torch.optim as optim
from PIL import Image

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class GestaltDataset(Dataset):
    def __init__(self, dataframe, data_dir, transform=None):
        self.dataframe = dataframe
        self.data_dir = data_dir
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        image_id = str(self.dataframe.iloc[idx, 0])  
        img_name = os.path.join(self.data_dir, f"{image_id}_aligned.jpg")
        image = Image.open(img_name)
        label = self.dataframe.iloc[idx, 2]  # Assuming label is in the 2nd column

        if self.transform:
            image = self.transform(image)

        return image, label
        
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])  

train_dataset = GestaltDataset(filtered_train, data_dir, transform=transform)
val_dataset = GestaltDataset(filtered_val, data_dir, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False) 
```

```python
model = models.resnet50(pretrained=True)
#model.classifier[1] = nn.Linear(model.classifier[1].in_features, 15)
model.fc = nn.Linear(model.fc.in_features, 15)
model = model.to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
num_epochs = 100

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(train_loader):.4f}")

    # Validation Loop
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    print(f"Validation Loss: {val_loss/len(val_loader):.4f}, Accuracy: {100 * correct / total:.2f}%")

print("Training Completed!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import (
    confusion_matrix, classification_report, accuracy_score, precision_score, 
    recall_score, f1_score, top_k_accuracy_score
)

def evaluate_and_visualize(model, test_loader, criterion, device, num_images=5):
    model.eval()  # Set the model to evaluation mode
    test_loss = 0.0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []
    all_probs = []

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            
            # Forward pass
            outputs = model(images)
            loss = criterion(outputs, labels)
            test_loss += loss.item()

            # Get predictions
            _, predicted = torch.max(outputs, 1)
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

            # Store probabilities for top-5 accuracy
            all_probs.extend(outputs.cpu().numpy())
            
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    avg_loss = test_loss / len(test_loader)
    accuracy = 100 * correct / total
    print(f"Test Loss: {avg_loss:.4f}, Test Accuracy: {accuracy:.2f}%")

    # Calculate metrics
    calculate_metrics(all_labels, all_preds, all_probs)
    visualize_predictions(test_loader, model, device, num_images)
    return avg_loss, accuracy


def calculate_metrics(true_labels, pred_labels, probabilities, num_classes=15):
    # Confusion Matrix
    cm = confusion_matrix(true_labels, pred_labels)
    print("\nConfusion Matrix:")
    print(cm)

    # Overall Metrics
    overall_accuracy = accuracy_score(true_labels, pred_labels)
    overall_precision = precision_score(true_labels, pred_labels, average='macro')
    overall_recall = recall_score(true_labels, pred_labels, average='macro')
    overall_f1 = f1_score(true_labels, pred_labels, average='macro')
    top5_accuracy = top_k_accuracy_score(true_labels, probabilities, k=5, labels=range(num_classes))

    print(f"\nOverall Metrics:")
    print(f"Accuracy: {overall_accuracy:.4f}")
    print(f"Macro Precision: {overall_precision:.4f}")
    print(f"Macro Recall: {overall_recall:.4f}")
    print(f"Macro F1-Score: {overall_f1:.4f}")
    print(f"Top-5 Accuracy: {top5_accuracy:.4f}")

    # Class-wise Metrics
    class_report = classification_report(true_labels, pred_labels, target_names=[f"Class {i}" for i in range(num_classes)])
    print("\nClass-wise Metrics:")
    print(class_report)

    # Class-wise support
    support = np.bincount(true_labels, minlength=num_classes)
    print(f"\nClass-wise Support: {support}")


def visualize_predictions(test_loader, model, device, num_images=5):
    model.eval()  # Ensure model is in eval mode
    images_shown = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

            for i in range(len(images)):
                if images_shown >= num_images:
                    break
                img = images[i].cpu().permute(1, 2, 0)  # Convert to HWC format for plotting
                plt.figure(figsize=(8, 4))
                plt.imshow(img)
                plt.title(f"Predicted: {predicted[i].item()}, Actual: {labels[i].item()}")
                plt.axis('off')
                plt.show()
                images_shown += 1
    plot_confusion_matrix(all_labels, all_preds)

def plot_confusion_matrix(true_labels, pred_labels):
    cm = confusion_matrix(true_labels, pred_labels)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=range(15), yticklabels=range(15))
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.show()
```

```python
import cv2
import mediapipe as mp
import pandas as pd
import numpy as np
import os

mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=1, min_detection_confidence=0.5)

def get_landmarks(img_path):
    import mediapipe as mp
    mp_face_mesh = mp.solutions.face_mesh
    face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=1)

    img = cv2.imread(img_path)
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(rgb_img)

    if results.multi_face_landmarks:
        landmarks = results.multi_face_landmarks[0].landmark
        # Extract x and y coordinates and scale them to the image dimensions
        return [(lm.x * img.shape[1], lm.y * img.shape[0]) for lm in landmarks]

    return []

# Function to calculate all required anthropometric distances
def calculate_distances(landmarks):
    distances = []
    landmarks = np.array(landmarks)
    # Eye to Eye Distance (outer corners)
    distances.append(np.linalg.norm(landmarks[336] - landmarks[130]))

    # Nose to Chin Distance (nose tip to chin tip)
    distances.append(np.linalg.norm(landmarks[168] - landmarks[152]))

    # Eye to Nose Tip (center of eyes to nose tip)
    eye_center = (landmarks[336] + landmarks[130]) / 2
    distances.append(np.linalg.norm(eye_center - landmarks[168]))

    # Nose Width (nasal bridge to nasal tip)
    distances.append(np.linalg.norm(landmarks[199] - landmarks[168]))

    # Mouth Width (outer corners of lips)
    distances.append(np.linalg.norm(landmarks[311] - landmarks[13]))

    # Distance from Nose Tip to Upper Lip (nose tip to upper lip)
    distances.append(np.linalg.norm(landmarks[168] - landmarks[13]))

    # Chin Width (outermost points of chin)
    distances.append(np.linalg.norm(landmarks[158] - landmarks[164]))

    # Jaw Width (outermost points of jawline)
    distances.append(np.linalg.norm(landmarks[10] - landmarks[334]))

    # Distance from Nose Base to Cheekbone (nose tip to cheekbones)
    distances.append(np.linalg.norm(landmarks[168] - landmarks[145]))
    distances.append(np.linalg.norm(landmarks[168] - landmarks[359]))

    # Distance from the top of the head to the chin (hairline to chin tip)
    distances.append(np.linalg.norm(landmarks[10] - landmarks[152]))

    # Additional measurements can be added similarly...

    return distances

df_landmarks = pd.DataFrame(columns=['image_id', 'distances', 'label'])

for dataset in [filtered_train, filtered_test, filtered_val]:
  for idx, row in dataset.iterrows():
    img_path = os.path.join(data_dir, f"{row['image_id']}_aligned.jpg")
    label = row['label']

    landmarks = get_landmarks(img_path)
    if landmarks is not None:
      distances = calculate_distances(landmarks)
      # Use concat instead of append
      df_landmarks = pd.concat([df_landmarks, pd.DataFrame({
          'image_id': [row['image_id']],
          'distances': [distances],
          'label': [label]
      })], ignore_index=True)

# Save to CSV
df_landmarks.to_csv('landmarks_distances.csv', index=False)
print(df_landmarks.head())
```
